```python
import sys
import csv
import subprocess

# Open the CSV file
with open('0 - srg_list_compilation.csv', 'r') as file:
    csv_reader = csv.reader(file)
    
    for row in csv_reader:
        N, K, LAMBDA, MU = map(int, row)
        if N > 30:
            break
        subprocess.call(["/usr/local/bin/python3", "srg_to_nonlinear_pseudoboolean_regularity_lambda_mu.py", str(N), str(K), str(LAMBDA), str(MU)])
        subprocess.call(["/usr/local/bin/python3", "srg_to_linear_pseudoboolean_regularity_lambda_mu.py", str(N), str(K), str(LAMBDA), str(MU)])
```
