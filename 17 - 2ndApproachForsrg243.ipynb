```python
import numpy as np

H = np.array([[2, 2, 2, 1, 1, 0, 1, 0, 0, 0, 0],
              [2, 2, 1, 2, 0, 1, 0, 1, 0, 0, 0],
              [2, 1, 2, 0, 2, 1, 0, 0, 1, 0, 0],
              [2, 1, 0, 2, 1, 2, 0, 0, 0, 1, 0],
              [2, 0, 1, 1, 2, 2, 0, 0, 0, 0, 1]])

vectors = []

for i in range(H.shape[1]):
    vectors.append(H[:, i])

H = (3 - H) % 3
for i in range(H.shape[1]):
    vectors.append(H[:, i])
main_vecs = vectors.copy()
print(main_vecs)
print(len(vectors))
for i in range(11):
    for j in range(i):
        vec = vectors[i] + vectors[j]
        vec = (vec + 6) % 3
        vectors.append(vec)

        vec = vectors[i] - vectors[j]
        vec = (vec + 6) % 3
        vectors.append(vec)

        vec = vectors[j] - vectors[i]
        vec = (vec + 6) % 3
        vectors.append(vec)

        vec = - vectors[i] - vectors[j]
        vec = (vec + 6) % 3
        vectors.append(vec)
vectors.append(vectors[0] - vectors[0])
```

```python
n = 243
adj = np.zeros((n, n), dtype=int)
for i in range(len(vectors)):
    v = vectors[i]
    for mv in main_vecs:
        neighbour = (v + mv) % 3
        indices = np.where((vectors == neighbour).all(axis=1))
        idn = indices[0]
        adj[i][idn] = 1
```

```python
def create_neighbor_sets(adj):
    n = len(adj)
    vertex_sets = [set() for _ in range(n)]

    for i in range(n):
        for j in range(n):
            if adj[i][j] == 1:
                vertex_sets[i].add(j)

    return vertex_sets

def find_shared_neighbors(vertex1, vertex2):
    neighbor_sets = create_neighbor_sets(adj)
    shared_neighbors = neighbor_sets[vertex1].intersection(neighbor_sets[vertex2])

    return list(shared_neighbors)

neighbor_sets = create_neighbor_sets(adj)


P9_vertices = [242, 1, 12, 2, 13, 30, 32, 31, 33]
for i in range(len(P9_vertices)):
    for j in range(i + 1, len(P9_vertices)):
        vertex1 = P9_vertices[i]
        vertex2 = P9_vertices[j]
        shared = find_shared_neighbors(vertex1, vertex2)
        shared_not_in_P9 = [neighbor for neighbor in shared if neighbor not in P9_vertices]
        print(f"Shared neighbors between vertex {vertex1} and vertex {vertex2}: {shared_not_in_P9}")
```

```python
def check_common_member(tuple1, tuple2):
    return any(x in tuple2 for x in tuple1)

T = len(triangles)
for i in range(T):
    for j in range(i + 1, T):
        l0 = check_common_member(triangles[i],triangles[j])

        if l0:
            continue
        # print(f"i:{i},j:{j}", triangles[i], triangles[j])
        ij_shared = 0
        for k in range(T):
            if k == i or k == j:
                continue
            l1 = check_common_member(triangles[k],triangles[i])
            l2 = check_common_member(triangles[k],triangles[j])
            if not l1 or not l2:
                continue
            ij_shared += 1
        if ij_shared > 3:
            print(f"WARNING! wrong answer {i}th and {j}th triangle", ij_shared)
```

```python
visited = set()
def dfs(v):
    global triangles, visited
    visited.add(v)
    for t in triangles:
        if v == t:
            continue
        if check_common_member(t, v) and t not in visited:
            dfs(t)

dfs(triangles[0])
print(len(visited))
```
