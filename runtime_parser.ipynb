```python
import os
import re

search_string = 'c CPU Time       :'

time_list = []
folder_path = 'solutions'
for file_name in os.listdir(folder_path):
    file_path = os.path.join(folder_path, file_name)
    N, K, LAMBDA, MU = [int(match) for match in re.findall(r'\d+', file_name)]
    file = open(file_path, 'r')
    for line in file:
        if search_string in line:
            cpu_time = line.split(':')[1].strip()
            time_list.append(((N,K,LAMBDA,MU), float(cpu_time[:-1])))
#             print(f"{N}, {K}, {LAMBDA}, {MU}: {cpu_time}")
time_list = sorted(time_list)
for x in time_list:
    N,K,LAMBDA,MU = x[0]
    cpu_time = x[1]
    print(f"{N}, {K}, {LAMBDA}, {MU}: {cpu_time}")
```
