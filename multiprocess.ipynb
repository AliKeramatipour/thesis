```python
import multiprocessing
import subprocess
import time
import queue
import csv

def process_file(filename):
    output_file_name = "solutions/" + filename[11:]
    output_file = open(output_file_name, "a")
    subprocess.call(["clasp", filename], stdout=output_file)

if __name__ == '__main__':
    # Create a queue of tasks
    filenames = []
    with open('srg_list_compilation.csv', 'r') as file:
        csv_reader = csv.reader(file)
        for row in csv_reader:
            N, K, LAMBDA, MU = map(int, row)
            if N > 30:
                continue
            filenames.append(f"non_linear/{N}K{K}L{LAMBDA}M{MU}.opb")

    pool = multiprocessing.Pool()
    pool.map_async(process_file, filenames)
    pool.close()
    pool.join()
```
